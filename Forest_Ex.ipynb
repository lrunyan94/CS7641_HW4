```python
import mdptoolbox, mdptoolbox.example
import matplotlib.pyplot as plt
```

```python
P, R = mdptoolbox.example.forest(S=1000, r1=1000, r2=500, p=0.001, is_sparse=False)
vi = mdptoolbox.mdp.ValueIteration(P, R, epsilon=0.01,discount=.8, max_iter=1000)
#vi.setVerbose()
stats = vi.run()
```

```python
vi.iter
```

```python
plt.plot(vi.V)
plt.ylabel("yName")
plt.xlabel("xName")
plt.title("dataName" + ": " +  "yName" + " vs " + "xName")
plt.show()
```
